```python
#Trying web API calls
import json
import requests
```

```python
api_url_base = 'https://data.cincinnati-oh.gov/resource/pid3-z56k.json'
headers = {'Content-Type': 'application/json'}
def get_account_info():
    response = requests.get(api_url_base, headers=headers)
    if response.status_code == 200:
        return json.loads(response.content.decode('utf-8'))
    else:
    return None

    
```

```python
account_info = get_account_info()

labels =['bid_project_number','status','project_name','requesting_department','buyer','procurement_type','inclusion_goal','date_due','bid_booklet_pdf','addendum_1_pdf']
for i in range(10):
    print("Record Number:",i)
    for label in labels:
        print(label,":",account_info[i][label])
```
